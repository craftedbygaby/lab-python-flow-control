```python
# Define a list called products that contains the following items: "t-shirt", "mug", "hat", "book", "keychain".
products = ['t-shirt', 'mug', 'hat', 'book', 'keychain']
```

```python
# Create an empty dictionary called inventory
inventory = {}
```

```python
# Adding inventory based on the list of products
for product in products:
    value = int(input(f'How many items of {product} is in inventory?'))
    inventory[product] = value
```

```python
print(inventory)
```

```python
# Create an empty set called customer_orders
customer_orders = set()
```

```python
# Function to input order from clients
def order():
    item_order = str(input('Which item client want to order?'))
    while item_order not in products:
        print('Invalid product. Write an item from products')
        item_order = str(input('Which item client want to order?'))
    customer_orders.add(item_order)
```

```python
# PENDING
# Ask the user to input the quantity of each product available in the inventory.
order()
order_more = str(input('Does the client wnat to order more? Y/N')).upper()

while order_more not in ['Y', 'N']:
    print('Invalid input. Please type Y or N')
    order_more = str(input('Does the client want to order more? Y/N ')).upper() 
    
while order_more == ('Y'):
    order()
    order_more = str(input('Does the client wnat to order more? Y/N')).upper()
    
if order_more == ('N'):
    print('Order Finished')  
    
print('Full customer order:', customer_orders)
```

```python
# Updated Inventory, only for the products that were ordered (those in "customer_orders")
for product in customer_orders:
    inventory[product] -= 1
```

```python
# printing inventory updated
for product, quantity in inventory.items():
    print(f'{product} : {quantity}')
```
